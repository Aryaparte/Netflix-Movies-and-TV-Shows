```python
import pandas as pd

# Load dataset
df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')

# 1. Identify and handle missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Example: Fill missing 'director' and 'cast' with "Unknown"
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)

# Drop rows with critical missing data (like missing title or date)
df.dropna(subset=['title', 'date_added'], inplace=True)

# 2. Remove duplicate rows
df.drop_duplicates(inplace=True)

# 3. Standardize text values
# Example: Standardize 'country' values (e.g., strip and lower)
df['country'] = df['country'].fillna('Unknown').str.strip().str.title()


country_map = {'Us': 'United States', 'Uk': 'United Kingdom'}
df['country'] = df['country'].replace(country_map)

# 4. Convert date formats to a consistent type
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

# 5. Rename column headers (lowercase, no spaces)
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

# 6. Check and fix data types
# 'release_year' should be int (already is), 'date_added' should be datetime (converted above)
print(df.dtypes)

# Final preview
print(df.head())
```
